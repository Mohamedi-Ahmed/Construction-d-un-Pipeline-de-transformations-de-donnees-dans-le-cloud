#Bronze Layer


##Chargement des secrets

```python
storage_account_name = dbutils.secrets.get(scope="azure-storage", key="storage-account-name")
storage_account_key = dbutils.secrets.get(scope="azure-storage", key="storage-account-key")
```

##Chargement des donn√©es brutes


```python
df_sql_bronze = spark.sql("SELECT * FROM `hive_metastore`.`default`.`bank_transactions_data`")
display(df_sql_bronze)
```

##Sauvegarde des donn√©es dans une table Bronze

```python
df_sql_bronze.write.format("delta").mode("overwrite").saveAsTable("bronze_bank_transactions")
```

##Sauvegarde des donn√©es dans un conteneur Azure

```python
spark.conf.set(
    f"fs.azure.account.key.{storage_account_name}.blob.core.windows.net",
    storage_account_key
)
```

```python
df_sql_bronze.write.format("delta").mode("overwrite").save(f"wasbs://bronze-data@{storage_account_name}.blob.core.windows.net/bronze")
```
